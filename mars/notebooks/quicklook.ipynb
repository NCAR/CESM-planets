```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors

import numpy as np
import xarray as xr
```

```python
run_dir = '~/cesm-dev/scratch/Mars_ne16_ne16_mg17_L49/run/'
file = 'Mars_ne16_ne16_mg17_L49.cam.h1.0001-01-0*-00000.nc'

ds = xr.open_mfdataset(run_dir+file)
```

```python
ds
```

```python
ds.T.mean(dim='lon').plot(y='lev', yscale='log', 
                          ylim=(7,1e-10),
                          col="time", col_wrap=4, 
                          robust=True)
plt.show()
```
