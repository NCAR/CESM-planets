```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors

import numpy as np
import xarray as xr
```

```python
# open the model history file
# note: if you have not installed nc-time-axis you may have to add 
# the decode_times=False argument to open_dataset or you can install it with
# conda install -c conda-forge nc-time-axis (or use mamba with same args)

run_dir = '/glade/scratch/pel/mars_ne16_ne16_mg17_short/run/'
file = 'mars_ne16_ne16_mg17_short.cam.h1.0001-01-01-00000.nc'

ds = xr.open_dataset(run_dir+file)
```

```python
ds.T.mean(dim='lon').plot(y='lev', yscale='log', 
                          ylim=(7,1e-10),
                          col="time", col_wrap=4, 
                          robust=True)
plt.show()
```

```python
ds.U.mean(dim='lon').plot(y='lev', yscale='log', 
                          ylim=(7,1e-10),
                          col="time", col_wrap=4, 
                          robust=True)
plt.show()
```

```python
ds.U.isel(lev=-1).plot(col="time", col_wrap=4, 
                          robust=True)
plt.show()
print('U (m/s) at surface:', np.round(ds.lev[-1].values,3), 'hPa')
```

```python
ds.U.sel(lev=1e-3, method='Nearest').plot(col="time", col_wrap=4, 
                          robust=True)
plt.show()
print(np.round(ds.lev.sel(lev=1e-3, method='Nearest').values,3), 'hPa')
```

```python
ds.U.sel(lev=1e-8, method='Nearest').plot(col="time", col_wrap=4, 
                          robust=True)
plt.show()
print(ds.lev.sel(lev=1e-8, method='Nearest').values, 'hPa')
```
