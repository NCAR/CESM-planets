```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
amp = np.loadtxt('amplitudes.dat')
```

```python
def plt_fit(idx, top, amp):
    levels = amp[:,0]
    
    plt.plot(amp[:,idx],levels, label=idx-1)
    x = amp[top:,0]
    y = amp[top:,idx]
    pc = np.polyfit(x, y, 5)
    p = np.poly1d(pc)
    plt.plot(p(x),x, label='fit')
    plt.legend()
    return pc
```

```python
for idx in np.arange(2,6):
    top = 9
    pc = plt_fit(idx, top, amp)
    p = np.flip(pc)
    print("%6.2e, %6.2e, %6.2e, %6.2e, %6.2e, %6.2e" % (p[0], p[1], p[2], p[3], p[4], p[5]))
plt.show()
```

```python
top = 9
pc = plt_fit(1, top, amp)
p = np.flip(pc)
print("%6.2e, %6.2e, %6.2e, %6.2e, %6.2e, %6.2e" % (p[0], p[1], p[2], p[3], p[4], p[5]))
plt.show()
```
